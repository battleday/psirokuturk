```python
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.patches as mpatches
```

```python
# figure wrappers and canvas drawing functions
def draw_square(rowSpan, colSpan, padding=0.02):
    """Takes a figure axis and position, and draws a rectangle there. Converts to matrix-style positioning"""
    points = [[colSpan[0] + padding, 1-rowSpan[1]+padding],
                      [colSpan[0] + padding, 1-rowSpan[0] - padding],
                      [colSpan[1] - padding, 1-rowSpan[0] - padding],
                      [colSpan[1] - padding, 1-rowSpan[1]+padding]] # first arguments are bottom left hand corner, then width, THEN height
    return points
    
def draw_circle(rowSpan, colSpan, fill, hatch, color='k', padding=0.02):
    """Takes a figure axis and position, and draws a circle there. Converts to matrix positioning"""
    c = plt.Circle((np.mean(colSpan), 1-np.mean(rowSpan)), 
                   radius=((rowSpan[1]-rowSpan[0])/2) - padding,
                  fill=fill, fc=color, hatch=hatch)
    return c

def draw_triangle(rowSpan, colSpan, padding=0.02):
    """Takes a figure axis and position, and draws a triangle there. Converts to matrix positioning"""
    top = 1-rowSpan[1] + padding
    bottom = 1-rowSpan[0] - padding
    points = [(colSpan[0]+padding, top),
    (np.mean(colSpan), bottom),
    (colSpan[1]-padding, top)]
          
    return points

def draw_inverted_triangle(rowSpan, colSpan, padding=0.02):
    """Takes a figure axis and position, and draws a triangle there. Converts to matrix positioning"""
    top = 1-rowSpan[0] - padding
    bottom = 1-rowSpan[1] + padding
    points = [(colSpan[0]+padding, top),
    (np.mean(colSpan), bottom),
    (colSpan[1]-padding, top)]

    return points

def draw_cross(rowSpan, colSpan, padding=0.02):
    """Takes a figure axis and position, and draws a triangle there. Converts to matrix positioning"""
    third_height = (rowSpan[1] - rowSpan[0] - 2*padding) / 3
    third_length = (colSpan[1] - colSpan[0] - 2*padding) / 3
    points = [(colSpan[0] + padding, 1-rowSpan[1] + padding + 2* third_height),
    (colSpan[0] + padding + third_length, 1-rowSpan[1] + padding + 2* third_height),
    (colSpan[0] + padding + third_length, 1-rowSpan[0] - padding),
    (colSpan[0] + padding + 2*third_length, 1-rowSpan[0] - padding),
    (colSpan[0] + padding + 2*third_length, 1-rowSpan[1] + padding + 2* third_height),
    (colSpan[0] + padding + 3*third_length, 1-rowSpan[1] + padding + 2* third_height),
    (colSpan[0] + padding + 3*third_length, 1-rowSpan[1] + padding + third_height),
    (colSpan[0] + padding + 2*third_length, 1-rowSpan[1] + padding + third_height),
    (colSpan[0] + padding + 2*third_length, 1-rowSpan[1] + padding),
    (colSpan[0] + padding + third_length, 1-rowSpan[1] + padding),
    (colSpan[0] + padding + third_length, 1-rowSpan[1] + padding + third_height),
    (colSpan[0] + padding, 1-rowSpan[1] + padding + third_height)]
    
    return points
    
def draw_ship(rowSpan, colSpan, padding=0.02):
    """Takes a figure axis and position, and draws a ship there. Converts to matrix positioning"""
    fourth_height = (rowSpan[1] - rowSpan[0] - 2*padding) / 4
    fifth_length = (colSpan[1] - colSpan[0] - 2*padding) / 5
    
    points = [(colSpan[0] + padding, 1 - rowSpan[1] + padding + 2*fourth_height),
    (colSpan[0] + fifth_length + padding, 1 - rowSpan[1] + padding + 2*fourth_height),
    (colSpan[0] + 2*fifth_length + padding, 1 - rowSpan[0] - padding),
    (colSpan[0] + 2*fifth_length + padding, 1 - rowSpan[1] + padding + 2*fourth_height),
    (colSpan[0] + 3*fifth_length + padding, 1 - rowSpan[1] + padding + 2*fourth_height),
    (colSpan[0] + 3*fifth_length + padding, 1 - rowSpan[0] - padding),
    (colSpan[0] + 4*fifth_length + padding, 1 - rowSpan[1] + padding + 2*fourth_height),
    (colSpan[0] + 5*fifth_length + padding, 1 - rowSpan[1] + padding + 2*fourth_height),
    (colSpan[0] + 3.5*fifth_length + padding, 1 - rowSpan[1] + padding),
    (colSpan[0] + 1.5*fifth_length + padding, 1 - rowSpan[1] + padding)]
    
    return points
    

def draw_gate(rowSpan, colSpan, padding=0.02):
    third_height = (rowSpan[1] - rowSpan[0] - 2*padding) / 3
    third_length = (colSpan[1] - colSpan[0] - 2*padding) / 3
    bottom = 1-rowSpan[1] + padding + (third_height/2)
    top = 1-rowSpan[0] - padding - (third_height/2)
    points = [(colSpan[0]+padding, bottom),
    (colSpan[0]+padding, top),
    (colSpan[1]-padding, top),
    (colSpan[1]-padding, bottom),
    (colSpan[1]-padding-third_length, bottom),
    (colSpan[1]-padding-third_length, bottom + third_height),
    (colSpan[0]+padding+third_length, bottom + third_height),
    (colSpan[0]+padding+third_length, bottom)]
    
    return points

def draw_stick(rowSpan, colSpan, padding=0.02):
    """Takes a figure axis and position, and draws a rectangle there. Converts to matrix-style positioning"""
    points= [(colSpan[0] + 2.5*padding, (1-rowSpan[1]+0.2*padding)),
                      (colSpan[0] + 2.5*padding, (1-rowSpan[0])-0.2*padding),
                      (colSpan[1] - 2.5*padding, (1-rowSpan[0])-0.2*padding),
                     (colSpan[1] - 2.5*padding, (1-rowSpan[1] + 0.2*padding))] # first arguments are bottom left hand corner, then width, THEN height
    return points
    


def draw_star(rowSpan, colSpan, padding=0.02):
    """Takes a figure axis and position, and draws a star there. Converts to matrix positioning"""
    third_height = (rowSpan[1] - rowSpan[0] - 2*padding) / 3
    third_length = (colSpan[1] - colSpan[0] - 2*padding) / 3
    points = [(colSpan[0] + padding + third_length, 1-rowSpan[1] + padding + 2* third_height),
    (colSpan[0] + padding + third_length, 1-rowSpan[0] - padding),
    (colSpan[0] + padding + 2*third_length, 1-rowSpan[1] + padding + 2* third_height),
    (colSpan[0] + padding + 3*third_length, 1-rowSpan[1] + padding + 2*third_height),
    (colSpan[0] + padding + 2*third_length, 1-rowSpan[1] + padding + third_height),
    (colSpan[0] + padding + 2*third_length, 1-rowSpan[1] + padding),
    (colSpan[0] + padding + third_length, 1-rowSpan[1] + padding + third_height),
    (colSpan[0] + padding, 1-rowSpan[1] + padding + third_height)]
    
    return points


def draw_human(rowSpan, colSpan, padding=0.02):
    """Takes a figure axis and position, and draws a humanoid sketch there. Converts to matrix positioning"""
    third_height = (rowSpan[1] - rowSpan[0] - 2*padding) / 3
    third_length = (colSpan[1] - colSpan[0] - 2*padding) / 3
    
    points = [(colSpan[0] + padding, 1-rowSpan[1] + padding + 3*third_height), # outer first
              (colSpan[0] + 2.3*padding, 1-rowSpan[1] + padding + 3*third_height), # inner fist
              (colSpan[0] + 1.9*padding, 1-rowSpan[1] + 3 * third_height), # top forearm
              (colSpan[0] + 1.9*padding, 1-rowSpan[1] + 2.8* third_height), # bottom left forearm
              (colSpan[0] + third_length, 1-rowSpan[1] + padding + 2.2*third_height), # bicep
              
    (colSpan[0] +1.5*third_length, 1-rowSpan[1] + padding + 2.2*third_height), # neck start left
              (colSpan[0] + 1.5*third_length, 1-rowSpan[1] + padding + 2.4*third_height), # neck finish left
              (colSpan[0] + 1.4*third_length, 1-rowSpan[1] + padding + 2.4*third_height), # head start left
    (colSpan[0] + padding + third_length, 1-rowSpan[0] - padding), # head top left
              
    (colSpan[1] - padding - third_length, 1-rowSpan[0] - padding), # head top right
    (colSpan[1] - padding - third_length, 1-rowSpan[1] + padding+ 2.4*third_height),# head finish right
              (colSpan[1]  - 1.5*third_length, 1-rowSpan[1] + padding + 2.4*third_height),#neck start right
              (colSpan[1] - 1.5*third_length, 1-rowSpan[1] + padding + 2.2*third_height), # neck finish right
              (colSpan[1] - third_length, 1-rowSpan[1] + padding + 2.2*third_height), # bicep right
              
              (colSpan[1] - 1.9*padding, 1-rowSpan[1] + 2.8* third_height), # bottom right forearm
              (colSpan[1] - 1.9*padding, 1-rowSpan[1] + 3 * third_height), # upper right arm
              (colSpan[1] - 2.3*padding, 1-rowSpan[1] + padding + 3*third_height), # inner right fist
    (colSpan[1] - padding, 1-rowSpan[1] + padding + 3*third_height), # outer right fist
    (colSpan[1] - 1.5*padding, 1-rowSpan[1] + padding + 2*third_height), # elbow right
              
    (colSpan[1] - padding - third_length, 1-rowSpan[1] + 2*third_height), # upper right ribs
    (colSpan[1] - padding - third_length, 1-rowSpan[1] + padding + third_height), # lower right ribs
    (colSpan[1] - 2*padding, 1-rowSpan[1] + third_height), # right knee
              (colSpan[1] - 2*padding, 1-rowSpan[1] + padding), # right outer foot
    (colSpan[1] - padding - third_length, 1-rowSpan[1] + padding), # right inner foot
              (colSpan[1] - padding - third_length, 1-rowSpan[1] + 2*padding), # right inner knee
    (colSpan[1] - padding - (3/2)*third_length, 1-rowSpan[1] + padding + (2/3)*third_length), # crotch
    (colSpan[0] + 2*padding + third_length, 1-rowSpan[1] + 1.5*padding),
    (colSpan[0] + 2*padding, 1-rowSpan[1] + 1.5*padding),
    (colSpan[0] + padding + third_length, 1-rowSpan[1] + padding + third_height),
    (colSpan[0] + padding + third_length, 1-rowSpan[1] + 2*third_height), # upper left ribs
    (colSpan[0] + 1.5*padding, 1-rowSpan[1] + padding + 2*third_height)] # elbow
    
    return points

def draw_x(rowSpan, colSpan, padding=0.02):
    """Takes a figure axis and position, and draws an x there. Converts to matrix positioning"""
    padding=2*padding
    top = 1-rowSpan[0] - padding
    bottom = 1-rowSpan[1] + padding
    middle = np.array((np.mean(colSpan), top - (top-bottom)/2))
    points = [(colSpan[0]+padding, top),
                middle, 
                  (colSpan[1]-padding, bottom),
                     middle,
               (colSpan[0]+padding, bottom),
              middle,
              (colSpan[1]-padding, top),
              middle
             ]
    return points

def draw_ninja(rowSpan, colSpan, padding=0.02):
    """Takes a figure axis and position, and draws a humanoid sketch there. Converts to matrix positioning"""
    third_height = (rowSpan[1] - rowSpan[0] - 2*padding) / 3
    third_length = (colSpan[1] - colSpan[0] - 2*padding) / 3
    
    points = [(colSpan[0] + 2.3*padding, 1-rowSpan[1] + padding + 1.5*third_height), # inner fist
             (colSpan[0] + padding, 1-rowSpan[1] + padding + 1.5*third_height), # outer first
              
              (colSpan[0] + 2.3*padding, 1-rowSpan[1] + padding + 2.2*third_height), # outer elbow
              (colSpan[0] + third_length, 1-rowSpan[1] + padding + 2.2*third_height), # bicep
              
    (colSpan[0] +1.5*third_length, 1-rowSpan[1] + padding + 2.2*third_height), # neck start left
              (colSpan[0] + 1.5*third_length, 1-rowSpan[1] + padding + 2.4*third_height), # neck finish left
              (colSpan[0] + 1.4*third_length, 1-rowSpan[1] + padding + 2.4*third_height), # head start left
    (colSpan[0] + padding + third_length, 1-rowSpan[0] - padding), # head top left
              
    (colSpan[1] - padding - third_length, 1-rowSpan[0] - padding), # head top right
    (colSpan[1] - padding - third_length, 1-rowSpan[1] + padding+ 2.4*third_height),# head finish right
              (colSpan[1]  - 1.5*third_length, 1-rowSpan[1] + padding + 2.4*third_height),#neck start right
              (colSpan[1] - 1.5*third_length, 1-rowSpan[1] + padding + 2.2*third_height), # neck finish right
              (colSpan[1] - third_length, 1-rowSpan[1] + padding + 2.2*third_height), # bicep right
              
              (colSpan[1] - 1.9*padding, 1-rowSpan[1] + 2.7* third_height), # bottom right forearm
              (colSpan[1] - 1.9*padding, 1-rowSpan[1] + 3 * third_height), # upper right arm
              (colSpan[1] - padding, 1-rowSpan[1] + padding + 2.8*third_height), # inner right fist
    (colSpan[1] - padding, 1-rowSpan[1] + padding + 2.5*third_height), # outer right fist
    (colSpan[1] - 1.6*padding, 1-rowSpan[1] + padding + 1.9*third_height), # elbow right
              
    (colSpan[1] - padding - third_length, 1-rowSpan[1] + 2.1*third_height), # upper right ribs
    (colSpan[1] - padding - third_length, 1-rowSpan[1] + padding + third_height), # lower right ribs
              
    (colSpan[1] - 2*padding, 1-rowSpan[1] + third_height), # right knee
              (colSpan[1] - 2*padding, 1-rowSpan[1] + padding), # right outer foot
    (colSpan[1] - padding - third_length, 1-rowSpan[1] + padding), # right inner foot
              (colSpan[1] - padding - third_length, 1-rowSpan[1] + 2*padding), # right inner knee
    (colSpan[1] - padding - (3/2)*third_length, 1-rowSpan[1] + padding + (2/3)*third_length), # crotch
              (colSpan[0] + padding + third_length, 1-rowSpan[1] + 2*padding), # left inner knee
    (colSpan[0] + padding + third_length, 1-rowSpan[1] + padding), # left inner foot
    (colSpan[0] + 2*padding, 1-rowSpan[1] + padding), # left outer foot
    (colSpan[0] + 2*padding, 1-rowSpan[1] + third_height), # left knee
              
    (colSpan[0] + padding + third_length, 1-rowSpan[1] + padding + third_height), # lower left ribs
    (colSpan[0] + padding + third_length, 1-rowSpan[1] + 2*third_height), # upper left ribs
    (colSpan[0] + 2.6*padding, 1-rowSpan[1] + 2.3*third_height)]# inner elbow 
    
    return points



def draw_figure(rowSpan, colSpan, padding=0.02):
    """Takes a figure axis and position, and draws a figure there. Converts to matrix positioning"""
    third_height = (rowSpan[1] - rowSpan[0] - 2*padding) / 3
    third_length = (colSpan[1] - colSpan[0] - 2*padding) / 3
    
    points = [(colSpan[0] + padding, 1-rowSpan[1] + padding + 3*third_height),
    (colSpan[0] + padding + third_length, 1-rowSpan[1] + padding + 2*third_height),
    (colSpan[0] + padding + third_length, 1-rowSpan[0] - padding),
    (colSpan[1] - padding - third_length, 1-rowSpan[0] - padding),
    (colSpan[1] - padding - third_length, 1-rowSpan[1] + padding + 2*third_height),
    (colSpan[1] - padding, 1-rowSpan[1] + padding + 3*third_height),
    (colSpan[1] - padding, 1-rowSpan[1] + padding + 2*third_height),
    (colSpan[1] - padding - third_length, 1-rowSpan[1] + padding + third_height),
    (colSpan[1] - padding - third_length, 1-rowSpan[1] + padding),
    (colSpan[0] + padding + third_length, 1-rowSpan[1] + padding),
    (colSpan[0] + padding + third_length, 1-rowSpan[1] + padding + third_height),
    (colSpan[0] + padding, 1-rowSpan[1] + padding + 2*third_height)]
    
    return points

def draw_canvas(height=10, width=10):
    """Wrapper function to return a figure"""
    fig = plt.figure(figsize=(width, height), 
                        facecolor='white', edgecolor='black')
    ax = fig.add_subplot(1, 1, 1)
    ax.axes.xaxis.set_visible(False)
    ax.axes.yaxis.set_visible(False)
    return fig, ax

def select_subregion(section, dims, nrows=5, ncols=5):
    """Splits each canvas into square rastered portions and returns limits of one of them.
    dims says whether there are an odd or even number of stimuli, and right shifts if even"""
    row_chunk_size = 1/nrows # think this is continuous
    col_chunk_size = 1/ncols
    row =  section // ncols
    col = section % ncols
    rowSpan = (row_chunk_size * row, row_chunk_size*(row+1))
    if dims == 'odd':
        colSpan = (col_chunk_size * col, col_chunk_size*(col+1))
    else:
        colSpan = (col_chunk_size * col + (0.5*col_chunk_size), col_chunk_size*(col+1) + (0.5*col_chunk_size))
    return rowSpan, colSpan

    
def output_stimulus(fig, ax, stimulus_list, dims='odd', save=False, save_path='.'):
    for fn, pos, fill, hatch in stimulus_list: # args must be ordered
        rS, cS = select_subregion(pos, dims=dims)
        try:
            points = fn(rS, cS)
            temp_patch = mpatches.Polygon(points, fill=fill, hatch=hatch, color='k', closed=True)
        except Exception as E:
            print(E)
            temp_patch = fn(rS, cS, fill=fill, hatch=hatch)
        ax.add_artist(temp_patch)

    if save:
        plt.savefig('{}.svg'.format(save_path), bbox_inches='tight')

# code for making full stimuli
def assemble_stimulus(stimList, dims, save, save_path):
    fig, ax = draw_canvas()
    output_stimulus(fig, ax, stimList, dims, save=save, save_path=save_path)
    plt.show()
    
```
